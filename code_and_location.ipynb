### 4. I have a dataframe with one (string) column and I'd like to split it into two (string) columns, with one column header as code and the other location.



```python
import pandas as pd
df = pd.DataFrame({"row" :['00000 UNITED STATES', '01000 ALABAMA',
                          '01001 Autauga County, AL', '01003 Baldwin County, AL',
                          '01005 Barbour County, AL']})
```

```python
df["code"] = df["row"].str[:6]
df["Location"] = df["row"].str[6:]
```

```python
df.drop("row", axis=1, inplace= True)
```
